```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
def rmse(y_true, y_pred) : 
    score = mean_squared_error(y_true, y_pred)**0.5
    return score

def mad(y_true, y_pred) : 
    score = mean_absolute_error(y_true, y_pred)/ len(y_pred)
    return score
```

```python
from sklearn.linear_model import Ridge

def ridge_selector(k) : 
    ridge_model = Ridge(alpha = k)
    ridge_model.fit(x_split,y_split)
    ridge_pred = ridge_model.predict(x_val)
    ridge_rmse = rmse(y_val,ridge_pred)
    return(ridge_rmse)
r_alphas = [0.00000005, 0.000000000000000000006, 0.005, 0.7, 1, 3, 5, 7, 10, 20, 40]
ridge_scores = []
```

```python
x = firedata_log[feature_list]
y = firedata_log[['area']]
x_train, x_test, y_train, y_test = train_test_split(x,y , test_size = 0.20, random_state = 42)
```

```python
import tensorflow.keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Dropout
from sklearn.model_selection import KFold
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
kf = KFold(n_splits=3, shuffle=True, random_state=41)
callbacks = [EarlyStopping(monitor='val_loss',patience=40,),
             ModelCheckpoint(filepath='best_weight.hdf5', verbose=1,
                             save_best_only=True, mode='auto')]
```

```python
i = 1
filepath='best_weight.hdf5'
ann_pred = np.zeros(len(x_test), dtype = np.float)
for train_index, valid_index in kf.split(x_train):
    X_train, X_valid = x_train.iloc[train_index, :], x_train.iloc[valid_index, :]
    Y_train, y_valid = y_train.values[train_index], y_train.values[valid_index]
    model = Sequential()
    model.add(Dense(64, input_dim=x_train.shape[1], activation = "tanh"))
    model.add(Dropout(0.5))
    model.add(Dense(16, activation = "relu"))
    model.add(Dropout(0.5))
    model.add(Dense(1, activation = "linear"))
    model.compile(loss="mean_squared_error", optimizer="adam")
    model.fit(X_train, Y_train, batch_size=256, epochs=1000, verbose=1,validation_data=(X_valid, y_valid), callbacks = callbacks)
    model.load_weights(filepath)
    preds_test_fold = model.predict(x_test,batch_size = 256, verbose=1)
    preds_test_fold = np.reshape(preds_test_fold, (96, ))
    ann_pred += preds_test_fold
    i = i+1
ann_pred /= 3
```

```python
ann_pred
```

```python
print('rmse : ', rmse(y_test, ann_pred))
print('mad : ', mad(y_test, ann_pred))
```
